```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
# NBA season we will be analyzing
year = 2019
# URL page we will scraping (see image above)
url = "https://www.basketball-reference.com/wnba/years/{}_per_game.html".format(year)
# this is the HTML from the given URL
html = urlopen(url)
soup = BeautifulSoup(html)
```

```python
result1 = soup.find_all('th', class_='left')
names = []
for result in result1:
    names.append(result.a.text)
names
#soup2 = BeautifulSoup(result, 'html.parser')
#soup2
```

```python
len(names)
```

```python
player_names = pd.DataFrame(names)

player_names
```

```python
player_names2 = player_names.reset_index()

player_names2
```

```python
# use findALL() to get the column headers
soup.findAll('tr', limit=2)
# use getText()to extract the text we need into a list
headers = [th.getText() for th in soup.findAll('tr', limit=2)[0].findAll('th')]
# exclude the first column as we will not need the ranking order from Basketball Reference for the analysis
headers = headers[1:]
headers
```

```python
# avoid the first header row
rows = soup.findAll('tr')[1:]
player_stats = [[td.getText() for td in rows[i].findAll('td')]
            for i in range(len(rows))]

rows
```

```python
stats = pd.DataFrame(player_stats, columns = headers)
stats2 = stats.dropna(how='all')
```

```python
stats3 = stats2.reset_index()
stats3
```

```python
stats4 = stats3.drop(columns=['index'])
stats4
```

```python
merge_table = pd.concat([player_names,stats4],axis=1)

merge_table2 = merge_table.rename(columns={0:"Player"})

merge_table2.head()
```

```python
teams = stats['Tm'].unique()

teams
```

```python
PHO = merge_table2.loc[(merge_table2["Tm"] == teams[1]), ["Player", "Tm", "G", "PTS"]]

PHO
```

```python
WNBA_teams_dict = {}
for team in teams:
    WNBA_teams_dict[team] = merge_table2.loc[(merge_table2["Tm"] == team), ["Player", "Tm", "G", "PTS"]]
    
WNBA_teams_dict['NYL']
```

```python
merge_table2.to_csv('WNBA_Data_Scrape_Final.csv',index=False)
```
